```python
items = divs[0].find_all("li")
```

```python
for li in items:
    label_span = li.find("span", class_="a-text-bold")
    if label_span:
        label = label_span.get_text(strip=True)
        label=label.split(" ")[0]
        print(label)
        if  label=="Manufacturer" :
            print("hi, ", label)
            print(label_span)
            value = label_span.find_next_sibling("span").get_text(strip=True)
            print(value)
           
```

```python
for li in items:
    spans = li.find_all("span")
    if len(spans) >= 2:
        label = spans[0].get_text(strip=True).split(":")[0].strip().split(" ")[0].split(" ")[0]
        value = spans[0].get_text(strip=True).split(":")[1].strip()
        print("hi 1")
        print("Label:", label)
        print("hi 2")
        print("Value:", value)
        print("hi 3")
        if label.strip() == "Manufacturer":
            print("Manufacturer value:", value)
```

```python
cnt =1
for item in new_soup.select('.a-unordered-list.a-nostyle.a-vertical.a-spacing-none.detail-bullet-list li'):
    # Find the label element
    label_span = item.find('span', class_='a-text-bold')
    if label_span:
        # Clean and normalize the label text
        label = label_span.get_text(strip=True).replace('\u200f', '').replace('\u200e', '').strip(':').strip()
        if 'Manufacturer' in label and cnt==5:
            # Extract the value from the next span
            value_span = item.find('span', class_=False)  # Assuming value span has no class
            if value_span:
                manufacturer = value_span.get_text(strip=True)
                print("Manufacturer:", manufacturer)
                break
        cnt=cnt+1
```

```python
ul = new_soup.find('ul', class_='a-unordered-list a-vertical a-spacing-mini')
about_items=[]
if ul:
    for li in ul.find_all('li', class_='a-spacing-mini'):
        item = li.find('span', class_='a-list-item')
        if item:
            # Clean and format the text
            text = item.get_text(strip=True)
            text = text.strip('"').replace('\n', ' ').strip()
            about_items.append(text)

# Combine items with double newlines
result = "\n\n".join(about_items)
print(result)
```

```python
for row in new_soup.select('tr[class*="po-"]'):
    label = row.find('span', class_='a-text-bold').get_text(strip=True)
    if 'Brand' in label:
        value = row.find('span', class_='po-break-word').get_text(strip=True)
        brand = value.split()[-1]  # Get last word assuming format "XXX Brand"
        print(brand)
```
